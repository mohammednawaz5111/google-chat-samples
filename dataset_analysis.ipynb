```python
# 4. Most Common Genres
plt.figure(figsize=(10, 6))
genre_counts = df_cleaned['genre'].value_counts().head(10)
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette='magma')
plt.title('Top 10 Most Common Genres')
plt.xlabel('Number of Movies')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

```python
# 5. Genre-wise Revenue Comparison (Inflation Adjusted)
plt.figure(figsize=(12, 6))
genre_revenue = df_cleaned.groupby('genre')['inflation_adjusted_gross'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=genre_revenue.values, y=genre_revenue.index, palette='cubehelix')
plt.title('Top 10 Genres by Total Inflation Adjusted Gross')
plt.xlabel('Total Inflation Adjusted Gross')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

```python
# 6. MPAA Rating Distribution
plt.figure(figsize=(8, 6))
rating_counts = df_cleaned['mpaa_rating'].value_counts()
sns.barplot(x=rating_counts.index, y=rating_counts.values, palette='pastel')
plt.title('Distribution of MPAA Ratings')
plt.xlabel('MPAA Rating')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.show()
```
